# Silver to gold - Promoting dimension tables

```python
import pyspark.sql.functions as F
catalog_name = 'edutrack'
```

## gld_dim_students

```python
df = spark.table('edutrack.silver.slv_students')
gld = df.select('student_id','name','department','year_of_study','date_of_birth','city','email')
gld.write.format('delta').mode('overwrite').option('mergeSchema','true').saveAsTable('edutrack.gold.gld_dim_students')
print('gld_dim_students:', spark.table('edutrack.gold.gld_dim_students').count(),'rows')
```

## gld_dim_courses

```python
df = spark.table('edutrack.silver.slv_courses')
gld = df.select('course_id','course_name','department','credits','level')
gld.write.format('delta').mode('overwrite').option('mergeSchema','true').saveAsTable('edutrack.gold.gld_dim_courses')
print('gld_dim_courses:', spark.table('edutrack.gold.gld_dim_courses').count(),'rows')
```
